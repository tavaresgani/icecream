```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ice = pd.read_csv('icecreamsales.csv')
ice
```

```python
celsius = (ice['Temperature'] - 32) * 5/9
ice['Temperature'] = celsius
ice #convertendo para celsius
```

```python
result = ice.sort_values('Temperature', ascending = False)
result #mostrando a temperatura da maior p menor
```

```python
plt.plot(result['Temperature'], result['Sales'], 'r.--', label='US$')

plt.title('Ice Cream Truck')

plt.xlabel('Temperatura (Celsius)')
plt.ylabel('Vendas (US$)')
plt.legend()
plt.savefig('icecreamsales3.pdf')
```
